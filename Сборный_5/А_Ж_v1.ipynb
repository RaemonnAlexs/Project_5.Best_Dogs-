Наш датасет содержит экспертные и краудсорсинговые оценки соответствия текста и изображения.

В файле с экспертными мнениями для каждой пары изображение-текст имеются оценки от трёх специалистов. Для решения задачи вы должны эти оценки агрегировать — превратить в одну. Существует несколько способов агрегации оценок, самый простой — голосование большинства: за какую оценку проголосовала большая часть экспертов (в нашем случае 2 или 3), та оценка и ставится как итоговая. Поскольку число экспертов меньше числа классов, может случиться, что каждый эксперт поставит разные оценки, например: 1, 4, 2. В таком случае данную пару изображение-текст можно исключить из датасета.

Вы можете воспользоваться другим методом агрегации оценок или придумать свой.

В файле с краудсорсинговыми оценками информация расположена в таком порядке:

1. Доля исполнителей, подтвердивших, что текст **соответствует** картинке.
2. Количество исполнителей, подтвердивших, что текст **соответствует** картинке.
3. Количество исполнителей, подтвердивших, что текст **не соответствует** картинке.

После анализа экспертных и краудсорсинговых оценок выберите либо одну из них, либо объедините их в одну по какому-то критерию: например, оценка эксперта принимается с коэффициентом 0.6, а крауда — с коэффициентом 0.4.

Ваша модель должна возвращать на выходе вероятность соответствия изображения тексту, поэтому целевая переменная должна иметь значения от 0 до 1.

<div class="alert alert-warning">

Решение превратить экспертные оценки в бинарный флаг `(sum >= 8)` — это довольно смелый, но рискованный шаг. На мой взгляд, это серьезная ошибка, которая «убивает» главную ценность данных.

**Почему бинаризация здесь вредна:**
Ценность экспертных оценок — именно в их способности передавать **нюансы и полутона**. Когда мы схлопываем шкалу от 1 до 4 в простой 0 и 1, мы теряем колоссальный объем информации.

**Давай посмотрим на примерах:**
Представь три пары «запрос-картинка» с разным весом релевантности:
* **Случай А (Слабое соответствие):** Эксперты поставили 2, 2 и 3. Сумма = 7. В твоей логике это **0**.
* **Случай Б (Хорошее соответствие):** Эксперты поставили 3, 3 и 2. Сумма = 8. В твоей логике это **1**.
* **Случай В (Идеальное соответствие):** Все поставили по 4. Сумма = 12. В твоей логике это тоже **1**.

Математически ты приравниваешь «неплохой» вариант (Случай Б) к «идеальному» (Случай В), хотя между ними огромная дистанция. При этом Случай А, который почти не отличается от Б, улетает в одну корзину с полным мусором. Модели будет очень сложно учиться на таких «рваных» данных, где $7.9 \approx 0$, а $8.1 \approx 1$.



**Как сделать лучше:**
Если ты решаешь задачу регрессии, оставайся в дробной шкале. Гораздо эффективнее использовать среднее арифметическое или медиану оценок экспертов:
$$\text{Score} = \frac{E_1 + E_2 + E_3}{3}$$
Это даст модели плавный градиент для обучения.

А приведение средней оценки экспертов к диапазону 0-1 можно решить простым линейным преобразованием `x-1/3``

**А если всё же хочется классификации?**
Если ты принципиально хочешь решать задачу классификации, то гораздо разумнее дополнительно использовать **крауд-оценки**. 
1.  **Их почти в 9 раз больше**, что критически важно для обучения нейронок (при соединении экспертных и крауд оценок получается около 50 тысяч размеченных пар.
2.  **Они бинарны по своей сути** (люди просто отвечают «да» или «нет» на вопрос о соответствии).

Оставь экспертам их «суперсилу» — точность и детализацию. Попробуй пересчитать таргет как среднее значение, и ты увидишь, что модель начнет гораздо тоньше чувствовать релевантность.

Лемматизируем текст для уменьшения вариаций слов в маске и первоначальной обработкой перед векторизацией текста.

````python
corpus = list(train_dataset['query_text'])
````

````python
def get_wordnet_pos(word):
    #Определяем POS-тег для WordNet
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize_english(text):
    #Лемматизация с учётом части речи
    lemmatizer = WordNetLemmatizer()
    words = text.split()
    lemmatized = []
    
    for word in words:
        pos = get_wordnet_pos(word)
        lemmatized.append(lemmatizer.lemmatize(word, pos))
    
    return ' '.join(lemmatized)

def clear_text(text):
    #Очистка текста
    res = re.sub(r"[^a-zA-Z']", ' ', text)
    c_text = " ".join(res.split())
    return c_text

def preprocess_text(text, remove_stopwords=False):
    #Полная предобработка
    text = text.lower()
    text = clear_text(text)
    text = lemmatize_english(text)
    
    if remove_stopwords:
        stop_words = set(stopwords.words('english'))
        words = text.split()
        words = [w for w in words if w not in stop_words]
        text = ' '.join(words)
    
    return text

# Проверка
print("Исходный текст:", corpus[15])
print()
print("Предобработанный (со стоп-словами):", preprocess_text(corpus[15], remove_stopwords=False))
print()
print("Предобработанный (без стоп-слов):", preprocess_text(corpus[15], remove_stopwords=True))
````

````python
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)  
model.eval()

def get_embeddings(texts, batch_size=32):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        encoded = tokenizer(texts[i:i+batch_size], padding=True, truncation=True, return_tensors='pt')
        with torch.no_grad():
            output = model(**encoded)
        
        mask = encoded['attention_mask'].unsqueeze(-1).float()
        token_emb = output.last_hidden_state * mask
        summed = token_emb.sum(dim=1)
        counts = mask.sum(dim=1).clamp(min=1e-9)
        batch_emb = summed / counts
        
        embeddings.append(batch_emb.numpy())
    
    return np.vstack(embeddings)

X_text = get_embeddings(train['p_text'].tolist(), batch_size=32)
print(X_text.shape)
````

```
from sklearn.model_selection import GroupShuffleSplit

train_indices, test_indices = next(gss.split(X=df.drop(columns=['target']), y=df['target'], groups=df['group_column']))
train_df, test_df = df.loc[train_indices], df.loc[test_indices]
```

````python
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
tokenizer = AutoTokenizer.from_pretrained(model_name)
minilm_model = AutoModel.from_pretrained(model_name)
minilm_model.eval();
````

````python
age_pattern = re.compile(
    r'\b(1[0-5]|[0-9])\s+(year|years|yr|yrs)\s+old\b|'
    r'\b(child|children|kid|kids|teen|teenager|teenagers|underage|minor|minors|boy|boys|girl|girls|young|'
    r'toddler|infant|baby|babies|preadolescent|adolescent|schoolboy|schoolgirl|schoolchild)\b',
    re.IGNORECASE
)
# Функция векторизации одного текста
def vectorize_text(text):
    encoded = tokenizer(text, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        output = minilm_model(**encoded)
    
    mask = encoded['attention_mask'].unsqueeze(-1).float()
    token_emb = output.last_hidden_state * mask
    summed = token_emb.sum(dim=1)
    counts = mask.sum(dim=1).clamp(min=1e-9)
    embedding = summed / counts
    
    return embedding.numpy().flatten()

# Функция process_query с проверкой возраста
def process_query(query_text, model_nn, image_vectors_dict, age_pattern=None):
    # Проверка на детский контент
    if age_pattern and age_pattern.search(str(query_text)):
        return None
    
    # Векторизуем текст запроса
    text_vector = vectorize_text(query_text)
    
    results = []
    for image_name, image_vector in image_vectors_dict.items():
        combined = np.concatenate([image_vector, text_vector])
        score = model_nn.predict(combined.reshape(1, -1), verbose=0)[0][0]
        results.append((query_text, score, image_name))
    
    results.sort(key=lambda x: x[1], reverse=True)
    
    if results:
        return results[0]
    return None

# Берём 10 случайных запросов
sample_queries = test_queries['query_text'].sample(
    min(10, len(test_queries)), random_state=42
)

image_folder = 'test_images'

# Обрабатываем каждый запрос
for idx, query_text in sample_queries.items():
    result = process_query(
        query_text=query_text,
        model_nn=model_nn,
        image_vectors_dict=image_vectors_dict,
        age_pattern=age_pattern
    )
    
    if result:
        query, score, image_name = result
        print(f"Запрос: {query}")
        print(f"Оценка: {score:.2f}")
        print(f"Изображение: {image_name}")
        
        # Вывод изображения
        img_path = os.path.join(image_folder, image_name)
        if os.path.exists(img_path):
            img = Image.open(img_path)
            plt.figure(figsize=(6, 6))
            plt.imshow(img)
            plt.title(f"Query: {query}\nScore: {score:.2f} | Image: {image_name}", fontsize=10)
            plt.axis('off')
            plt.show()
        else:
            print(f"⚠️ Файл не найден: {img_path}")
        
        print("-" * 50)
    else:
        print(f"Запрос '{query_text}' заблокирован")
        print("-" * 50)
````
