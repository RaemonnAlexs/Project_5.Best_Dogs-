Р’Р°Рј РҝРҫСҖСғСҮРөРҪРҫ СҖР°Р·СҖР°РұРҫСӮР°СӮСҢ РҙРөРјРҫРҪСҒСӮСҖР°СҶРёРҫРҪРҪСғСҺ РІРөСҖСҒРёСҺ РҝРҫРёСҒРәР° РёР·РҫРұСҖР°Р¶РөРҪРёР№ РҝРҫ Р·Р°РҝСҖРҫСҒСғ.

Р”Р»СҸ РҙРөРјРҫРҪСҒСӮСҖР°СҶРёРҫРҪРҪРҫР№ РІРөСҖСҒРёРё РҪСғР¶РҪРҫ РҫРұСғСҮРёСӮСҢ РјРҫРҙРөР»СҢ, РәРҫСӮРҫСҖР°СҸ РҝРҫР»СғСҮРёСӮ РІРөРәСӮРҫСҖРҪРҫРө РҝСҖРөРҙСҒСӮР°РІР»РөРҪРёРө РёР·РҫРұСҖР°Р¶РөРҪРёСҸ, РІРөРәСӮРҫСҖРҪРҫРө РҝСҖРөРҙСҒСӮР°РІР»РөРҪРёРө СӮРөРәСҒСӮР°, Р° РҪР° РІСӢС…РҫРҙРө РІСӢРҙР°СҒСӮ СҮРёСҒР»Рҫ РҫСӮ 0 РҙРҫ 1 вҖ” РҝРҫРәР°Р¶РөСӮ, РҪР°СҒРәРҫР»СҢРәРҫ СӮРөРәСҒСӮ Рё РәР°СҖСӮРёРҪРәР° РҝРҫРҙС…РҫРҙСҸСӮ РҙСҖСғРі РҙСҖСғРіСғ.

### РһРҝРёСҒР°РҪРёРө РҙР°РҪРҪСӢС…

Р”Р°РҪРҪСӢРө РҙРҫСҒСӮСғРҝРҪСӢ РҝРҫ [СҒСҒСӢР»РәРө](https://code.s3.yandex.net/datasets/dsplus_integrated_project_4.zip).

Р’ С„Р°Р№Р»Рө `train_dataset.csv` РҪР°С…РҫРҙРёСӮСҒСҸ РёРҪС„РҫСҖРјР°СҶРёСҸ, РҪРөРҫРұС…РҫРҙРёРјР°СҸ РҙР»СҸ РҫРұСғСҮРөРҪРёСҸ: РёРјСҸ С„Р°Р№Р»Р° РёР·РҫРұСҖР°Р¶РөРҪРёСҸ, РёРҙРөРҪСӮРёС„РёРәР°СӮРҫСҖ РҫРҝРёСҒР°РҪРёСҸ Рё СӮРөРәСҒСӮ РҫРҝРёСҒР°РҪРёСҸ. Р”Р»СҸ РҫРҙРҪРҫР№ РәР°СҖСӮРёРҪРәРё РјРҫР¶РөСӮ РұСӢСӮСҢ РҙРҫСҒСӮСғРҝРҪРҫ РҙРҫ 5 РҫРҝРёСҒР°РҪРёР№. РҳРҙРөРҪСӮРёС„РёРәР°СӮРҫСҖ РҫРҝРёСҒР°РҪРёСҸ РёРјРөРөСӮ С„РҫСҖРјР°СӮ `<РёРјСҸ С„Р°Р№Р»Р° РёР·РҫРұСҖР°Р¶РөРҪРёСҸ>#<РҝРҫСҖСҸРҙРәРҫРІСӢР№ РҪРҫРјРөСҖ РҫРҝРёСҒР°РҪРёСҸ>`.

Р’ РҝР°РҝРәРө `train_images` СҒРҫРҙРөСҖР¶Р°СӮСҒСҸ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ РҙР»СҸ СӮСҖРөРҪРёСҖРҫРІРәРё РјРҫРҙРөР»Рё.

Р’ С„Р°Р№Р»Рө `CrowdAnnotations.tsv` вҖ” РҙР°РҪРҪСӢРө РҝРҫ СҒРҫРҫСӮРІРөСӮСҒСӮРІРёСҺ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ Рё РҫРҝРёСҒР°РҪРёСҸ, РҝРҫР»СғСҮРөРҪРҪСӢРө СҒ РҝРҫРјРҫСүСҢСҺ РәСҖР°СғРҙСҒРҫСҖСҒРёРҪРіР°. РқРҫРјРөСҖР° РәРҫР»РҫРҪРҫРә Рё СҒРҫРҫСӮРІРөСӮСҒСӮРІСғСҺСүРёР№ СӮРёРҝ РҙР°РҪРҪСӢС…:

1. РҳРјСҸ С„Р°Р№Р»Р° РёР·РҫРұСҖР°Р¶РөРҪРёСҸ.
2. РҳРҙРөРҪСӮРёС„РёРәР°СӮРҫСҖ РҫРҝРёСҒР°РҪРёСҸ.
3. Р”РҫР»СҸ Р»СҺРҙРөР№, РҝРҫРҙСӮРІРөСҖРҙРёРІСҲРёС…, СҮСӮРҫ РҫРҝРёСҒР°РҪРёРө СҒРҫРҫСӮРІРөСӮСҒСӮРІСғРөСӮ РёР·РҫРұСҖР°Р¶РөРҪРёСҺ.
4. РҡРҫР»РёСҮРөСҒСӮРІРҫ СҮРөР»РҫРІРөРә, РҝРҫРҙСӮРІРөСҖРҙРёРІСҲРёС…, СҮСӮРҫ РҫРҝРёСҒР°РҪРёРө СҒРҫРҫСӮРІРөСӮСҒСӮРІСғРөСӮ РёР·РҫРұСҖР°Р¶РөРҪРёСҺ.
5. РҡРҫР»РёСҮРөСҒСӮРІРҫ СҮРөР»РҫРІРөРә, РҝРҫРҙСӮРІРөСҖРҙРёРІСҲРёС…, СҮСӮРҫ РҫРҝРёСҒР°РҪРёРө РҪРө СҒРҫРҫСӮРІРөСӮСҒСӮРІСғРөСӮ РёР·РҫРұСҖР°Р¶РөРҪРёСҺ.

Р’ С„Р°Р№Р»Рө `ExpertAnnotations.tsv` СҒРҫРҙРөСҖР¶Р°СӮСҒСҸ РҙР°РҪРҪСӢРө РҝРҫ СҒРҫРҫСӮРІРөСӮСҒСӮРІРёСҺ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ Рё РҫРҝРёСҒР°РҪРёСҸ, РҝРҫР»СғСҮРөРҪРҪСӢРө РІ СҖРөР·СғР»СҢСӮР°СӮРө РҫРҝСҖРҫСҒР° СҚРәСҒРҝРөСҖСӮРҫРІ. РқРҫРјРөСҖР° РәРҫР»РҫРҪРҫРә Рё СҒРҫРҫСӮРІРөСӮСҒСӮРІСғСҺСүРёР№ СӮРёРҝ РҙР°РҪРҪСӢС…:

1. РҳРјСҸ С„Р°Р№Р»Р° РёР·РҫРұСҖР°Р¶РөРҪРёСҸ.
2. РҳРҙРөРҪСӮРёС„РёРәР°СӮРҫСҖ РҫРҝРёСҒР°РҪРёСҸ.

3, 4, 5 вҖ” РҫСҶРөРҪРәРё СӮСҖС‘С… СҚРәСҒРҝРөСҖСӮРҫРІ.

РӯРәСҒРҝРөСҖСӮСӢ СҒСӮР°РІСҸСӮ РҫСҶРөРҪРәРё РҝРҫ СҲРәР°Р»Рө РҫСӮ 1 РҙРҫ 4, РіРҙРө 1 вҖ” РёР·РҫРұСҖР°Р¶РөРҪРёРө Рё Р·Р°РҝСҖРҫСҒ СҒРҫРІРөСҖСҲРөРҪРҪРҫ РҪРө СҒРҫРҫСӮРІРөСӮСҒСӮРІСғСҺСӮ РҙСҖСғРі РҙСҖСғРіСғ, 2 вҖ” Р·Р°РҝСҖРҫСҒ СҒРҫРҙРөСҖР¶РёСӮ СҚР»РөРјРөРҪСӮСӢ РҫРҝРёСҒР°РҪРёСҸ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ, РҪРҫ РІ СҶРөР»РҫРј Р·Р°РҝСҖРҫСҒ СӮРөРәСҒСӮСғ РҪРө СҒРҫРҫСӮРІРөСӮСҒСӮРІСғРөСӮ, 3 вҖ” Р·Р°РҝСҖРҫСҒ Рё СӮРөРәСҒСӮ СҒРҫРҫСӮРІРөСӮСҒСӮРІСғСҺСӮ СҒ СӮРҫСҮРҪРҫСҒСӮСҢСҺ РҙРҫ РҪРөРәРҫСӮРҫСҖСӢС… РҙРөСӮР°Р»РөР№, 4 вҖ” Р·Р°РҝСҖРҫСҒ Рё СӮРөРәСҒСӮ СҒРҫРҫСӮРІРөСӮСҒСӮРІСғСҺСӮ РҝРҫР»РҪРҫСҒСӮСҢСҺ.

Р’ С„Р°Р№Р»Рө `test_queries.csv` РҪР°С…РҫРҙРёСӮСҒСҸ РёРҪС„РҫСҖРјР°СҶРёСҸ, РҪРөРҫРұС…РҫРҙРёРјР°СҸ РҙР»СҸ СӮРөСҒСӮРёСҖРҫРІР°РҪРёСҸ: РёРҙРөРҪСӮРёС„РёРәР°СӮРҫСҖ Р·Р°РҝСҖРҫСҒР°, СӮРөРәСҒСӮ Р·Р°РҝСҖРҫСҒР° Рё СҖРөР»РөРІР°РҪСӮРҪРҫРө РёР·РҫРұСҖР°Р¶РөРҪРёРө. Р”Р»СҸ РҫРҙРҪРҫР№ РәР°СҖСӮРёРҪРәРё РјРҫР¶РөСӮ РұСӢСӮСҢ РҙРҫСҒСӮСғРҝРҪРҫ РҙРҫ 5 РҫРҝРёСҒР°РҪРёР№. РҳРҙРөРҪСӮРёС„РёРәР°СӮРҫСҖ РҫРҝРёСҒР°РҪРёСҸ РёРјРөРөСӮ С„РҫСҖРјР°СӮ `<РёРјСҸ С„Р°Р№Р»Р° РёР·РҫРұСҖР°Р¶РөРҪРёСҸ>#<РҝРҫСҖСҸРҙРәРҫРІСӢР№ РҪРҫРјРөСҖ РҫРҝРёСҒР°РҪРёСҸ>`.

Р’ РҝР°РҝРәРө `test_images` СҒРҫРҙРөСҖР¶Р°СӮСҒСҸ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ РҙР»СҸ СӮРөСҒСӮРёСҖРҫРІР°РҪРёСҸ РјРҫРҙРөР»Рё.

##  РҳСҒСҒР»РөРҙРҫРІР°СӮРөР»СҢСҒРәРёР№ Р°РҪР°Р»РёР· РҙР°РҪРҪСӢС…

РқР°СҲ РҙР°СӮР°СҒРөСӮ СҒРҫРҙРөСҖР¶РёСӮ СҚРәСҒРҝРөСҖСӮРҪСӢРө Рё РәСҖР°СғРҙСҒРҫСҖСҒРёРҪРіРҫРІСӢРө РҫСҶРөРҪРәРё СҒРҫРҫСӮРІРөСӮСҒСӮРІРёСҸ СӮРөРәСҒСӮР° Рё РёР·РҫРұСҖР°Р¶РөРҪРёСҸ.

Р’ С„Р°Р№Р»Рө СҒ СҚРәСҒРҝРөСҖСӮРҪСӢРјРё РјРҪРөРҪРёСҸРјРё РҙР»СҸ РәР°Р¶РҙРҫР№ РҝР°СҖСӢ РёР·РҫРұСҖР°Р¶РөРҪРёРө-СӮРөРәСҒСӮ РёРјРөСҺСӮСҒСҸ РҫСҶРөРҪРәРё РҫСӮ СӮСҖС‘С… СҒРҝРөСҶРёР°Р»РёСҒСӮРҫРІ. Р”Р»СҸ СҖРөСҲРөРҪРёСҸ Р·Р°РҙР°СҮРё РІСӢ РҙРҫР»Р¶РҪСӢ СҚСӮРё РҫСҶРөРҪРәРё Р°РіСҖРөРіРёСҖРҫРІР°СӮСҢ вҖ” РҝСҖРөРІСҖР°СӮРёСӮСҢ РІ РҫРҙРҪСғ. РЎСғСүРөСҒСӮРІСғРөСӮ РҪРөСҒРәРҫР»СҢРәРҫ СҒРҝРҫСҒРҫРұРҫРІ Р°РіСҖРөРіР°СҶРёРё РҫСҶРөРҪРҫРә, СҒР°РјСӢР№ РҝСҖРҫСҒСӮРҫР№ вҖ” РіРҫР»РҫСҒРҫРІР°РҪРёРө РұРҫР»СҢСҲРёРҪСҒСӮРІР°: Р·Р° РәР°РәСғСҺ РҫСҶРөРҪРәСғ РҝСҖРҫРіРҫР»РҫСҒРҫРІР°Р»Р° РұРҫР»СҢСҲР°СҸ СҮР°СҒСӮСҢ СҚРәСҒРҝРөСҖСӮРҫРІ (РІ РҪР°СҲРөРј СҒР»СғСҮР°Рө 2 РёР»Рё 3), СӮР° РҫСҶРөРҪРәР° Рё СҒСӮР°РІРёСӮСҒСҸ РәР°Рә РёСӮРҫРіРҫРІР°СҸ. РҹРҫСҒРәРҫР»СҢРәСғ СҮРёСҒР»Рҫ СҚРәСҒРҝРөСҖСӮРҫРІ РјРөРҪСҢСҲРө СҮРёСҒР»Р° РәР»Р°СҒСҒРҫРІ, РјРҫР¶РөСӮ СҒР»СғСҮРёСӮСҢСҒСҸ, СҮСӮРҫ РәР°Р¶РҙСӢР№ СҚРәСҒРҝРөСҖСӮ РҝРҫСҒСӮР°РІРёСӮ СҖР°Р·РҪСӢРө РҫСҶРөРҪРәРё, РҪР°РҝСҖРёРјРөСҖ: 1, 4, 2. Р’ СӮР°РәРҫРј СҒР»СғСҮР°Рө РҙР°РҪРҪСғСҺ РҝР°СҖСғ РёР·РҫРұСҖР°Р¶РөРҪРёРө-СӮРөРәСҒСӮ РјРҫР¶РҪРҫ РёСҒРәР»СҺСҮРёСӮСҢ РёР· РҙР°СӮР°СҒРөСӮР°.

Р’СӢ РјРҫР¶РөСӮРө РІРҫСҒРҝРҫР»СҢР·РҫРІР°СӮСҢСҒСҸ РҙСҖСғРіРёРј РјРөСӮРҫРҙРҫРј Р°РіСҖРөРіР°СҶРёРё РҫСҶРөРҪРҫРә РёР»Рё РҝСҖРёРҙСғРјР°СӮСҢ СҒРІРҫР№.

Р’ С„Р°Р№Р»Рө СҒ РәСҖР°СғРҙСҒРҫСҖСҒРёРҪРіРҫРІСӢРјРё РҫСҶРөРҪРәР°РјРё РёРҪС„РҫСҖРјР°СҶРёСҸ СҖР°СҒРҝРҫР»РҫР¶РөРҪР° РІ СӮР°РәРҫРј РҝРҫСҖСҸРҙРәРө:

1. Р”РҫР»СҸ РёСҒРҝРҫР»РҪРёСӮРөР»РөР№, РҝРҫРҙСӮРІРөСҖРҙРёРІСҲРёС…, СҮСӮРҫ СӮРөРәСҒСӮ **СҒРҫРҫСӮРІРөСӮСҒСӮРІСғРөСӮ** РәР°СҖСӮРёРҪРәРө.

Р’ РҪРөРәРҫСӮРҫСҖСӢС… СҒСӮСҖР°РҪР°С…, РіРҙРө СҖР°РұРҫСӮР°РөСӮ РІР°СҲР° РәРҫРјРҝР°РҪРёСҸ, РҙРөР№СҒСӮРІСғСҺСӮ РҫРіСҖР°РҪРёСҮРөРҪРёСҸ РҝРҫ РҫРұСҖР°РұРҫСӮРәРө РёР·РҫРұСҖР°Р¶РөРҪРёР№: РҝРҫРёСҒРәРҫРІСӢРј СҒРөСҖРІРёСҒР°Рј Рё СҒРөСҖРІРёСҒР°Рј, РҝСҖРөРҙРҫСҒСӮР°РІР»СҸСҺСүРёРј РІРҫР·РјРҫР¶РҪРҫСҒСӮСҢ РҝРҫРёСҒРәР°, Р·Р°РҝСҖРөСүРөРҪРҫ РұРөР· СҖР°Р·СҖРөСҲРөРҪРёСҸ СҖРҫРҙРёСӮРөР»РөР№ РёР»Рё Р·Р°РәРҫРҪРҪСӢС… РҝСҖРөРҙСҒСӮР°РІРёСӮРөР»РөР№ РҝСҖРөРҙРҫСҒСӮР°РІР»СҸСӮСҢ Р»СҺРұСғСҺ РёРҪС„РҫСҖРјР°СҶРёСҺ, РІ СӮРҫРј СҮРёСҒР»Рө, РҪРҫ РҪРө РёСҒРәР»СҺСҮРёСӮРөР»СҢРҪРҫ СӮРөРәСҒСӮСӢ, РёР·РҫРұСҖР°Р¶РөРҪРёСҸ, РІРёРҙРөРҫ Рё Р°СғРҙРёРҫ, СҒРҫРҙРөСҖР¶Р°СүРёРө РҫРҝРёСҒР°РҪРёРө, РёР·РҫРұСҖР°Р¶РөРҪРёРө РёР»Рё Р·Р°РҝРёСҒСҢ РіРҫР»РҫСҒР° РҙРөСӮРөР№. Р РөРұС‘РҪРәРҫРј СҒСҮРёСӮР°РөСӮСҒСҸ Р»СҺРұРҫР№ СҮРөР»РҫРІРөРә, РҪРө РҙРҫСҒСӮРёРіСҲРёР№ 16 Р»РөСӮ.

Р’ РІР°СҲРөРј СҒРөСҖРІРёСҒРө СҒСӮСҖРҫРіРҫ СҒР»РөРҙСғСҺСӮ Р·Р°РәРҫРҪР°Рј СҒСӮСҖР°РҪ, РІ РәРҫСӮРҫСҖСӢС… СҖР°РұРҫСӮР°СҺСӮ. РҹРҫСҚСӮРҫРјСғ РҝСҖРё РҝРҫРҝСӢСӮРәРө РҝРҫСҒРјРҫСӮСҖРөСӮСҢ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ, Р·Р°РҝСҖРөСүС‘РҪРҪСӢРө Р·Р°РәРҫРҪРҫРҙР°СӮРөР»СҢСҒСӮРІРҫРј, РІРјРөСҒСӮРҫ РәР°СҖСӮРёРҪРҫРә РҝРҫРәР°Р·СӢРІР°РөСӮСҒСҸ РҙРёСҒРәР»РөР№РјРөСҖ:

> This image is unavailable in your country in compliance with local laws
>

РһРҙРҪР°РәРҫ Сғ РІР°СҒ РІ PoC РҪРөСӮ РІРҫР·РјРҫР¶РҪРҫСҒСӮРё РІРҫСҒРҝРҫР»СҢР·РҫРІР°СӮСҢСҒСҸ РҙР°РҪРҪСӢРј С„СғРҪРәСҶРёРҫРҪР°Р»РҫРј. РҹРҫСҚСӮРҫРјСғ РІСҒРө РёР·РҫРұСҖР°Р¶РөРҪРёСҸ, РәРҫСӮРҫСҖСӢРө РҪР°СҖСғСҲР°СҺСӮ РҙР°РҪРҪСӢР№ Р·Р°РәРҫРҪ, РҪСғР¶РҪРҫ СғРҙР°Р»РёСӮСҢ РёР· РҫРұСғСҮР°СҺСүРөР№ РІСӢРұРҫСҖРәРё.


РӣРөРјРјР°СӮРёР·РёСҖСғРөРј СӮРөРәСҒСӮ РҙР»СҸ СғРјРөРҪСҢСҲРөРҪРёСҸ РІР°СҖРёР°СҶРёР№ СҒР»РҫРІ РІ РјР°СҒРәРө Рё РҝРөСҖРІРҫРҪР°СҮР°Р»СҢРҪРҫР№ РҫРұСҖР°РұРҫСӮРәРҫР№ РҝРөСҖРөРҙ РІРөРәСӮРҫСҖРёР·Р°СҶРёРөР№ СӮРөРәСҒСӮР°.

````python
corpus = list(train_dataset['query_text'])
````

````python
def get_wordnet_pos(word):
    #РһРҝСҖРөРҙРөР»СҸРөРј POS-СӮРөРі РҙР»СҸ WordNet
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize_english(text):
    #РӣРөРјРјР°СӮРёР·Р°СҶРёСҸ СҒ СғСҮС‘СӮРҫРј СҮР°СҒСӮРё СҖРөСҮРё
    lemmatizer = WordNetLemmatizer()
    words = text.split()
    lemmatized = []
    
    for word in words:
        pos = get_wordnet_pos(word)
        lemmatized.append(lemmatizer.lemmatize(word, pos))
    
    return ' '.join(lemmatized)

def clear_text(text):
    #РһСҮРёСҒСӮРәР° СӮРөРәСҒСӮР°
    res = re.sub(r"[^a-zA-Z']", ' ', text)
    c_text = " ".join(res.split())
    return c_text

def preprocess_text(text, remove_stopwords=False):
    #РҹРҫР»РҪР°СҸ РҝСҖРөРҙРҫРұСҖР°РұРҫСӮРәР°
    text = text.lower()
    text = clear_text(text)
    text = lemmatize_english(text)
    
    if remove_stopwords:
        stop_words = set(stopwords.words('english'))
        words = text.split()
        words = [w for w in words if w not in stop_words]
        text = ' '.join(words)
    
    return text

# РҹСҖРҫРІРөСҖРәР°
print("РҳСҒС…РҫРҙРҪСӢР№ СӮРөРәСҒСӮ:", corpus[15])
print()
print("РҹСҖРөРҙРҫРұСҖР°РұРҫСӮР°РҪРҪСӢР№ (СҒРҫ СҒСӮРҫРҝ-СҒР»РҫРІР°РјРё):", preprocess_text(corpus[15], remove_stopwords=False))
print()
print("РҹСҖРөРҙРҫРұСҖР°РұРҫСӮР°РҪРҪСӢР№ (РұРөР· СҒСӮРҫРҝ-СҒР»РҫРІ):", preprocess_text(corpus[15], remove_stopwords=True))
````

````python
sentence = "Beautiful flowers were growing in the big gardens. He went to the store for his mother"
result = preprocess_text(sentence, remove_stopwords=True)
print(result)
print("beautiful flower grow big garden go store mother")
````

````python
# РҹРҫР»СғСҮРөРҪРҪСӢР№ СҖРөР·СғР»СҢСӮР°СӮ РҝСҖРёРјРөРҪСҸРөРј РәРҫ РІСҒРөРј СӮРөРәСҒСӮР°Рј
train_dataset['p_text'] = train_dataset['query_text'].progress_apply(preprocess_text, remove_stopwords=True)
````

Р’СҒРө РёР·РҫРұСҖР°Р¶РөРҪРёСҸ, РәРҫСӮРҫСҖСӢРө РҪР°СҖСғСҲР°СҺСӮ РҙР°РҪРҪСӢР№ Р·Р°РәРҫРҪ,  СғРҙР°Р»РөРҪСӢ - 1553 Р·Р°РҝСҖРҫСҒРҫРІ. РһРұСғСҮР°СҺСүР°СҸ РІСӢРұРҫСҖРәР° РіРҫСӮРҫРІР° Рә РҙР°Р»СҢРҪРөР№СҲРөР№ СҖР°РұРҫСӮРө.

##  Р’РөРәСӮРҫСҖРёР·Р°СҶРёСҸ РёР·РҫРұСҖР°Р¶РөРҪРёР№

РҹРөСҖРөР№РҙС‘Рј Рә РІРөРәСӮРҫСҖРёР·Р°СҶРёРё РёР·РҫРұСҖР°Р¶РөРҪРёР№.

РЎР°РјСӢР№ РҝСҖРёРјРёСӮРёРІРҪСӢР№ СҒРҝРҫСҒРҫРұ вҖ” РҝСҖРҫСҮРөСҒСӮСҢ РёР·РҫРұСҖР°Р¶РөРҪРёРө Рё РҝСҖРөРІСҖР°СӮРёСӮСҢ РҝРҫР»СғСҮРөРҪРҪСғСҺ РјР°СӮСҖРёСҶСғ РІ РІРөРәСӮРҫСҖ. РўР°РәРҫР№ СҒРҝРҫСҒРҫРұ РҪР°Рј РҪРө РҝРҫРҙС…РҫРҙРёСӮ: РҙР»РёРҪР° РІРөРәСӮРҫСҖРҫРІ РјРҫР¶РөСӮ РұСӢСӮСҢ СҒРёР»СҢРҪРҫ СҖР°Р·РҪРҫР№, СӮР°Рә РәР°Рә СҖР°Р·РјРөСҖСӢ РёР·РҫРұСҖР°Р¶РөРҪРёР№ СҖР°Р·РҪСӢРө. РҹРҫСҚСӮРҫРјСғ СҒСӮРҫРёСӮ РҫРұСҖР°СӮРёСӮСҢСҒСҸ Рә СҒРІС‘СҖСӮРҫСҮРҪСӢРј СҒРөСӮСҸРј: РҫРҪРё РҝРҫР·РІРҫР»СҸСҺСӮ "РІСӢРҙРөР»РёСӮСҢ" РіР»Р°РІРҪСӢРө РәРҫРјРҝРҫРҪРөРҪСӮСӢ РёР·РҫРұСҖР°Р¶РөРҪРёР№. РҡР°Рә СҚСӮРҫ СҒРҙРөР»Р°СӮСҢ? РқСғР¶РҪРҫ РІСӢРұСҖР°СӮСҢ РәР°РәСғСҺ-Р»РёРұРҫ Р°СҖС…РёСӮРөРәСӮСғСҖСғ, РҪР°РҝСҖРёРјРөСҖ ResNet-18, РҝРҫСҒРјРҫСӮСҖРөСӮСҢ РҪР° СҒР»РҫРё Рё РёСҒРәР»СҺСҮРёСӮСҢ РҝРҫР»РҪРҫСҒРІСҸР·РҪСӢРө СҒР»РҫРё, РәРҫСӮРҫСҖСӢРө РҫСӮРІРөСҮР°СҺСӮ Р·Р° РәРҫРҪРөСҮРҪРҫРө РҝСҖРөРҙСҒРәР°Р·Р°РҪРёРө. РҹСҖРё СҚСӮРҫРј РјРҫР¶РҪРҫ Р·Р°РіСҖСғР·РёСӮСҢ РјРҫРҙРөР»СҢ РҙР°РҪРҪРҫР№ Р°СҖС…РёСӮРөРәСӮСғСҖСӢ, РҝСҖРөРҙРІР°СҖРёСӮРөР»СҢРҪРҫ РҪР°СӮСҖРөРҪРёСҖРҫРІР°РҪРҪСғСҺ РҪР° РҙР°СӮР°СҒРөСӮРө ImageNet.

````python
# Р—Р°РіСҖСғР·РәР° РҝСҖРөРҙРҫРұСғСҮРөРҪРҪРҫР№ РјРҫРҙРөР»Рё ResNet-18
resnet18 = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)

# РЈРҙР°Р»СҸРөРј РҝРҫСҒР»РөРҙРҪРёР№ РҝРҫР»РҪРҫСҒРІСҸР·РҪСӢР№ СҒР»РҫР№
resnet18 = torch.nn.Sequential(*list(resnet18.children())[:-1])
resnet18.eval()

# РўСҖР°РҪСҒС„РҫСҖРјР°СҶРёРё РҙР»СҸ РёР·РҫРұСҖР°Р¶РөРҪРёР№
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
````

````python
# РӨСғРҪРәСҶРёСҸ РҝРҫР»СғСҮРөРҪРёСҸ РІРөРәСӮРҫСҖР°
def get_image_vector(img_path):
    img = Image.open(img_path).convert('RGB')
    img_tensor = transform(img).unsqueeze(0)
    with torch.no_grad():
        features = resnet18(img_tensor)
    return features.squeeze().numpy()

# Р’РөРәСӮРҫСҖРёР·Р°СҶРёСҸ РІ РҝРҫСҖСҸРҙРәРө train['image']
image_folder = 'train_images'
vectors = []
missing_images = []

for img_name in train['image']:
    img_path = os.path.join(image_folder, img_name)
    if os.path.exists(img_path):
        vector = get_image_vector(img_path)
        vectors.append(vector)
    else:
        vectors.append(np.zeros(512))  # РёР»Рё None, РёР»Рё РҝСҖРҫРҝСғСҒСӮРёСӮСҢ
        missing_images.append(img_name)
        print(f"РқРө РҪР°Р№РҙРөРҪРҫ: {img_name}")

vectors = np.array(vectors)
print(f"Р Р°Р·РјРөСҖ РјР°СӮСҖРёСҶСӢ РІРөРәСӮРҫСҖРҫРІ: {vectors.shape}")  # (4269, 512)
````

## Р’РөРәСӮРҫСҖРёР·Р°СҶРёСҸ СӮРөРәСҒСӮРҫРІ

РЎР»РөРҙСғСҺСүРёР№ СҚСӮР°Рҝ вҖ” РІРөРәСӮРҫСҖРёР·Р°СҶРёСҸ СӮРөРәСҒСӮРҫРІ. Р’СӢ РјРҫР¶РөСӮРө РҝРҫСҚРәСҒРҝРөСҖРёРјРөРҪСӮРёСҖРҫРІР°СӮСҢ СҒ РҪРөСҒРәРҫР»СҢРәРёРјРё СҒРҝРҫСҒРҫРұР°РјРё РІРөРәСӮРҫСҖРёР·Р°СҶРёРё СӮРөРәСҒСӮРҫРІ:

- tf-idf
- word2vec
- \*СӮСҖР°РҪСҒС„РҫСҖРјРөСҖСӢ (РҪР°РҝСҖРёРјРөСҖ Bert)

\* вҖ” РөСҒР»Рё РІСӢ РёР·СғСҮР°Р»Рё СӮСҖР°РҪСҒС„РҫСҖРјРөСҖСӢ РІ СҒРҝСҖРёРҪСӮРө РңР°СҲРёРҪРҪРҫРө РҫРұСғСҮРөРҪРёРө РҙР»СҸ СӮРөРәСҒСӮРҫРІ.


````python
# Р’РөРәСӮРҫСҖРёР·Р°СҶРёСҸ СӮРөРәСҒСӮРҫРІ СҒ РҝРҫРјРҫСүСҢСҺ TF-IDF
tfidf = TfidfVectorizer(max_features=5000)  # РјРҫР¶РҪРҫ РҫРіСҖР°РҪРёСҮРёСӮСҢ РәРҫР»РёСҮРөСҒСӮРІРҫ РҝСҖРёР·РҪР°РәРҫРІ

# РһРұСғСҮР°РөРј РІРөРәСӮРҫСҖР°Р№Р·РөСҖ РҪР° СӮРөРәСҒСӮР°С… Рё РҝСҖРөРҫРұСҖР°Р·СғРөРј
X_text = tfidf.fit_transform(train['p_text'])

print(f"Р Р°Р·РјРөСҖ РјР°СӮСҖРёСҶСӢ: {X_text.shape}")  # (РәРҫР»РёСҮРөСҒСӮРІРҫ СҒСӮСҖРҫРә, РәРҫР»РёСҮРөСҒСӮРІРҫ РҝСҖРёР·РҪР°РәРҫРІ)
print(f"РҡРҫР»РёСҮРөСҒСӮРІРҫ СғРҪРёРәР°Р»СҢРҪСӢС… СҒР»РҫРІ: {len(tfidf.get_feature_names_out())}")
````

##  РһРұСҠРөРҙРёРҪРөРҪРёРө РІРөРәСӮРҫСҖРҫРІ

РҹРҫРҙРіРҫСӮРҫРІСҢСӮРө РҙР°РҪРҪСӢРө РҙР»СҸ РҫРұСғСҮРөРҪРёСҸ: РҫРұСҠРөРҙРёРҪРёСӮРө РІРөРәСӮРҫСҖСӢ РёР·РҫРұСҖР°Р¶РөРҪРёР№ Рё РІРөРәСӮРҫСҖСӢ СӮРөРәСҒСӮРҫРІ СҒ СҶРөР»РөРІРҫР№ РҝРөСҖРөРјРөРҪРҪРҫР№.

````python
# РҹСҖРөРҫРұСҖР°Р·СғРөРј СҖР°Р·СҖРөР¶РөРҪРҪСғСҺ РјР°СӮСҖРёСҶСғ РІ numpy array
X_text_array = X_text.toarray()

# РһРұСҠРөРҙРёРҪСҸРөРј СҒ РІРөРәСӮРҫСҖР°РјРё РёР·РҫРұСҖР°Р¶РөРҪРёР№
X_combined = np.hstack([vectors, X_text_array])  # vectors РёР· РҝСҖРөРҙСӢРҙСғСүРөРіРҫ Р·Р°РҙР°РҪРёСҸ
print(f"Р Р°Р·РјРөСҖ РҫРұСҠРөРҙРёРҪРөРҪРҪРҫР№ РјР°СӮСҖРёСҶСӢ: {X_combined.shape}")
````

##  РһРұСғСҮРөРҪРёРө РјРҫРҙРөР»Рё РҝСҖРөРҙСҒРәР°Р·Р°РҪРёСҸ СҒРҫРҫСӮРІРөСӮСҒСӮРІРёСҸ

Р”Р»СҸ РҫРұСғСҮРөРҪРёСҸ СҖР°Р·РҙРөР»РёСӮРө РҙР°СӮР°СҒРөСӮ РҪР° СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСғСҺ Рё СӮРөСҒСӮРҫРІСғСҺ РІСӢРұРҫСҖРәРё. РҹСҖРҫСҒСӮРҫРө СҒР»СғСҮР°Р№РҪРҫРө СҖР°Р·РұРёРөРҪРёРө РҪРө РҝРҫРҙС…РҫРҙРёСӮ: РҪСғР¶РҪРҫ РёСҒРәР»СҺСҮРёСӮСҢ РҝРҫРҝР°РҙР°РҪРёРө РёР·РҫРұСҖР°Р¶РөРҪРёСҸ Рё РІ РҫРұСғСҮР°СҺСүСғСҺ, Рё РІ СӮРөСҒСӮРҫРІСғСҺ РІСӢРұРҫСҖРәРё.
Р”Р»СҸ СӮРҫРіРҫ СҮСӮРҫРұСӢ СғСҮРөСҒСӮСҢ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ РҝСҖРё СҖР°Р·РұРёРөРҪРёРё, РјРҫР¶РҪРҫ РІРҫСҒРҝРҫР»СҢР·РҫРІР°СӮСҢСҒСҸ РәР»Р°СҒСҒРҫРј [GroupShuffleSplit](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GroupShuffleSplit.html) РёР· РұРёРұР»РёРҫСӮРөРәРё sklearn.model_selection.

РҡРҫРҙ РҪРёР¶Рө СҖР°Р·РұРёРІР°РөСӮ РҙР°СӮР°СҒРөСӮ РҪР° СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСғСҺ Рё СӮРөСҒСӮРҫРІСғСҺ РІСӢРұРҫСҖРәРё РІ РҝСҖРҫРҝРҫСҖСҶРёРё 7:3 СӮР°Рә, СҮСӮРҫ СҒСӮСҖРҫРәРё СҒ РҫРҙРёРҪР°РәРҫРІСӢРј Р·РҪР°СҮРөРҪРёРөРј 'group_column' РұСғРҙСғСӮ СҒРҫРҙРөСҖР¶Р°СӮСҢСҒСҸ Р»РёРұРҫ РІ СӮРөСҒСӮРҫРІРҫРј, Р»РёРұРҫ РІ СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪРҫРј РҙР°СӮР°СҒРөСӮРө.

```
from sklearn.model_selection import GroupShuffleSplit
gss = GroupShuffleSplit(n_splits=1, train_size=.7, random_state=42)
train_indices, test_indices = next(gss.split(X=df.drop(columns=['target']), y=df['target'], groups=df['group_column']))
train_df, test_df = df.loc[train_indices], df.loc[test_indices]

```

РҡР°РәСғСҺ РјРҫРҙРөР»СҢ РёСҒРҝРҫР»СҢР·РҫРІР°СӮСҢ вҖ” РІСӢРұРөСҖРёСӮРө СҒР°РјРҫСҒСӮРҫСҸСӮРөР»СҢРҪРҫ. РўР°РәР¶Рө РІР°Рј РҝСҖРөРҙСҒСӮРҫРёСӮ РІСӢРұСҖР°СӮСҢ РјРөСӮСҖРёРәСғ РәР°СҮРөСҒСӮРІР° Р»РёРұРҫ СҖРөР°Р»РёР·РҫРІР°СӮСҢ СҒРІРҫСҺ.

````python
y = train['expert_binary']

gss = GroupShuffleSplit(n_splits=1, train_size=.7, random_state=42)

train_indices, test_indices = next(gss.split(X = X_combined, y = y, groups= train['image']))

X_train, X_test = X_combined[train_indices], X_combined[test_indices]
y_train, y_test = y.iloc[train_indices], y.iloc[test_indices]
````

Р’СӢРұРөСҖРөРј RandomForestRegressor РҙР»СҸ СҖР°РұРҫСӮСӢ СҒ СҖР°Р·СҖРөР¶РөРҪРҪСӢРјРё РҙР°РҪРҪСӢРјРё РұРөР· РјР°СҒСҲСӮР°РұРёСҖРҫРІР°РҪРёСҸ. Р’РөРәСӮРҫСҖСӢ РёР·РҫРұСҖР°Р¶РөРҪРёР№ РёРјРөСҺСӮ Р·РҪР°СҮРөРҪРёСҸ РҫСӮ 0 РҙРҫ 1. TF-IDF РІРөРәСӮРҫСҖСӢ РҫСӮ 0 Рё РұРҫР»РөРө 1 . RandomForestRegressor СғСҒСӮРҫР№СҮРёРІ Рә РјСғР»СҢСӮРёРәРҫР»Р»РёРҪРөР°СҖРҪРҫСҒСӮРё, РҪРө СӮСҖРөРұСғРөСӮ РҪРҫСҖРјР°Р»РёР·Р°СҶРёРё СҖР°СҒРҝСҖРөРҙРөР»РөРҪРёСҸ, СҚС„С„РөРәСӮРёРІРөРҪ РҝСҖРё РұРҫСҢР»СҲРҫРј РәРҫР»РёСҮРөСҒСӮРІРө РҝСҖРёР·РҪР°РәРҫРІ.

model = RandomForestRegressor(
    n_estimators=150,           # РәРҫР»РёСҮРөСҒСӮРІРҫ РҙРөСҖРөРІСҢРөРІ (РұРҫР»СҢСҲРө = СҒСӮР°РұРёР»СҢРҪРөРө)
    max_depth=30,               # РіР»СғРұРёРҪР° РҙРөСҖРөРІР° (РҫРіСҖР°РҪРёСҮРёРІР°РөРј РҝРөСҖРөРҫРұСғСҮРөРҪРёРө)
    min_samples_split=5,       # РјРёРҪ. РҫРұСҠРөРәСӮРҫРІ РҙР»СҸ СҖР°Р·РұРёРөРҪРёСҸ СғР·Р»Р°
    min_samples_leaf=2,         # РјРёРҪ. РҫРұСҠРөРәСӮРҫРІ РІ Р»РёСҒСӮРө
    max_features='sqrt',        # РәРҫР»-РІРҫ РҝСҖРёР·РҪР°РәРҫРІ РҙР»СҸ СҖР°Р·РұРёРөРҪРёСҸ (sqrt(1461) вүҲ 38)
    n_jobs=-1,                  
    random_state=42                             
)
model.fit(X_train, y_train)

````python
model_nn = Sequential([
    Input(shape=(X_train.shape[1],)),
    
    # РЎР»РҫР№ 1
    Dense(256, activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.4),
    
    # РЎР»РҫР№ 2
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.4),
    
    # РЎР»РҫР№ 3
    Dense(64, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.3),
    
    # РЎР»РҫР№ 4
    Dense(32, activation='relu'),
    Dropout(0.2),
    
    # Р’СӢС…РҫРҙ
    Dense(1, activation='sigmoid')
])

model_nn.compile(
    optimizer=Adam(learning_rate=0.0005),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        verbose=1
    )
]

history = model_nn.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=callbacks
)

y_pred_train = model_nn.predict(X_train).flatten()
y_pred_test = model_nn.predict(X_test).flatten()

# РңРөСӮСҖРёРәРё
y_pred_binary_train = (y_pred_train > 0.5).astype(int)
y_pred_binary_test = (y_pred_test > 0.5).astype(int)

train_acc = accuracy_score(y_train, y_pred_binary_train)
test_acc = accuracy_score(y_test, y_pred_binary_test)
train_auc = roc_auc_score(y_train, y_pred_train)
test_auc = roc_auc_score(y_test, y_pred_test)


print()
print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy:  {test_acc:.4f}")
print(f"Train ROC-AUC:  {train_auc:.4f}")
print(f"Test ROC-AUC:   {test_auc:.4f}")
````

Р’СӢРұРөСҖРөРј 2 РјРөСӮСҖРёРәРё: Accuracy - РҙР»СҸ РҝСҖРҫСҒСӮРҫСӮСӢ Рё РҪР°РіР»СҸРҙРҪРҫСҒСӮРё, ROC-AUC - РҙР»СҸ РҫСҶРөРҪРәРё РәР°СҮРөСҒСӮРІР° РјРҫРҙРөР»Рё РҪРөР·Р°РІРёСҒРёРјРҫ РҫСӮ РҝРҫСҖРҫРіР°.

y_pred = model.predict(X_test)

y_pred_binary = (y_pred > 0.5).astype(int)

accuracy = accuracy_score(y_test, y_pred_binary)
roc_auc = roc_auc_score(y_test, y_pred)

print(f"Accuracy:  {accuracy:.4f}")

##  РўРөСҒСӮРёСҖРҫРІР°РҪРёРө РјРҫРҙРөР»Рё

РқР°СҒСӮР°Р»Рҫ РІСҖРөРјСҸ РҝСҖРҫСӮРөСҒСӮРёСҖРҫРІР°СӮСҢ РјРҫРҙРөР»СҢ. Р”Р»СҸ СҚСӮРҫРіРҫ РҝРҫР»СғСҮРёСӮРө СҚРјРұРөРҙРҙРёРҪРіРё РҙР»СҸ РІСҒРөС… СӮРөСҒСӮРҫРІСӢС… РёР·РҫРұСҖР°Р¶РөРҪРёР№ РёР· РҝР°РҝРәРё `test_images`, РІСӢРұРөСҖРёСӮРө СҒР»СғСҮР°Р№РҪСӢРө 10 Р·Р°РҝСҖРҫСҒРҫРІ РёР· С„Р°Р№Р»Р° `test_queries.csv` Рё РҙР»СҸ РәР°Р¶РҙРҫРіРҫ Р·Р°РҝСҖРҫСҒР° РІСӢРІРөРҙРёСӮРө РҪР°РёРұРҫР»РөРө СҖРөР»РөРІР°РҪСӮРҪРҫРө РёР·РҫРұСҖР°Р¶РөРҪРёРө. РЎСҖР°РІРҪРёСӮРө РІРёР·СғР°Р»СҢРҪРҫ РәР°СҮРөСҒСӮРІРҫ РҝРҫРёСҒРәР°.

````python
# РҹРҫР»СғСҮР°РөРј 100 СғРҪРёРәР°Р»СҢРҪСӢС… РёР·РҫРұСҖР°Р¶РөРҪРёР№
unique_images = test_queries['image'].unique()
print(f"Р—Р°РіСҖСғР¶Р°РөРј {len(unique_images)} СғРҪРёРәР°Р»СҢРҪСӢС… РёР·РҫРұСҖР°Р¶РөРҪРёР№...")

# РҹСҖРөРҙРІСӢСҮРёСҒР»РөРҪРёРө РІРөРәСӮРҫСҖРҫРІ РёР·РҫРұСҖР°Р¶РөРҪРёР№
image_vectors_dict = {}
for img_name in unique_images:
    img_path = os.path.join(image_folder, img_name)
    if os.path.exists(img_path):
        image_vectors_dict[img_name] = get_image_vector(img_path)
    else:
        image_vectors_dict[img_name] = np.zeros(512)
        print(f"РқРө РҪР°Р№РҙРөРҪРҫ: {img_name}")

print("Р’СҒРө РёР·РҫРұСҖР°Р¶РөРҪРёСҸ РІРөРәСӮРҫСҖРёР·РҫРІР°РҪСӢ!")

# ============================================
# 2. РӨРЈРқРҡРҰРҳРҜ РҹРһРҳРЎРҡРҗ РҳР—РһР‘Р РҗР–Р•РқРҳРҜ РҹРһ РўР•РҡРЎРўРЈ
# ============================================
def find_image_by_text(query_text):
    """
    РӨСғРҪРәСҶРёСҸ РҝРҫРёСҒРәР° РёР·РҫРұСҖР°Р¶РөРҪРёСҸ РҝРҫ СӮРөРәСҒСӮРҫРІРҫРјСғ Р·Р°РҝСҖРҫСҒСғ
    
    РҹР°СҖР°РјРөСӮСҖСӢ:
    - query_text: str, СӮРөРәСҒСӮРҫРІСӢР№ Р·Р°РҝСҖРҫСҒ
    
    Р’РҫР·РІСҖР°СүР°РөСӮ:
    - best_image: str, РёРјСҸ Р»СғСҮСҲРөРіРҫ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ
    - best_score: float, РҫСҶРөРҪРәР° СҒРҫРҫСӮРІРөСӮСҒСӮРІРёСҸ (0-1)
    - is_valid: bool, True РөСҒР»Рё Р·Р°РҝСҖРҫСҒ РәРҫСҖСҖРөРәСӮРөРҪ (РҪРө РҝСҖРҫ РҙРөСӮРөР№)
    """
    
    # РЁР°Рі 1: РҹСҖРҫРІРөСҖРәР° РҪР° Р·Р°РҝСҖРөСүРөРҪРҪСӢРө Р·Р°РҝСҖРҫСҒСӢ (РҙРөСӮРё РҙРҫ 16 Р»РөСӮ)
    age_pattern = r'\b(1[0-5]|[0-9])\s+(year|years)\s+old\b|\b(child|children|kid|kids|teen|teenager|underage|minor|boy|girl|young)\b'
    
    if re.search(age_pattern, query_text, re.IGNORECASE):
        print("вҡ пёҸ Р”РёСҒРәР»РөР№РјРөСҖ: РҳР·РҫРұСҖР°Р¶РөРҪРёРө РҪРөРҙРҫСҒСӮСғРҝРҪРҫ РІ РІР°СҲРөР№ СҒСӮСҖР°РҪРө РІ СҒРҫРҫСӮРІРөСӮСҒСӮРІРёРё СҒ РјРөСҒСӮРҪСӢРјРё Р·Р°РәРҫРҪР°РјРё")
        return None, 0.0, False
    
    # РЁР°Рі 2: РӣРөРјРјР°СӮРёР·Р°СҶРёСҸ СӮРөРәСҒСӮР°
    processed_text = preprocess_text(query_text, remove_stopwords=True)
    
    # РЁР°Рі 3: Р’РөРәСӮРҫСҖРёР·Р°СҶРёСҸ СӮРөРәСҒСӮР°
    text_vec = tfidf.transform([processed_text]).toarray()
    
    # РЁР°Рі 4: РҹРҫРёСҒРә Р»СғСҮСҲРөРіРҫ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ СҒСҖРөРҙРё 100 Р·Р°СҖР°РҪРөРө РІРөРәСӮРҫСҖРёР·РҫРІР°РҪРҪСӢС…
    best_score = -1
    best_image = None
    
    for img_name, img_vector in image_vectors_dict.items():
        X = np.hstack([img_vector.reshape(1, -1), text_vec])
        score = model_nn.predict(X)[0].item()
        if score > best_score:
            best_score = score
            best_image = img_name
    
    return best_image, best_score, True

# ============================================
# 3. Р”Р•РңРһРқРЎРўР РҗРҰРҳРҜ: 10 РЎРӣРЈР§РҗРҷРқР«РҘ Р—РҗРҹР РһРЎРһР’
# ============================================
print("\n" + "="*70)
print("РҹРһРҳРЎРҡ РҳР—РһР‘Р РҗР–Р•РқРҳРҷ РҹРһ РўР•РҡРЎРўРһР’Р«Рң Р—РҗРҹР РһРЎРҗРң")
print("="*70)

# РҹРҫР»СғСҮР°РөРј 10 СҒР»СғСҮР°Р№РҪСӢС… СғРҪРёРәР°Р»СҢРҪСӢС… Р·Р°РҝСҖРҫСҒРҫРІ РёР· СӮРөСҒСӮРҫРІРҫРіРҫ РҙР°СӮР°СҒРөСӮР°
sample_queries = test_queries.drop_duplicates('query_text').sample(10, random_state=33)

for idx, row in sample_queries.iterrows():
    query_text = row['query_text']
    
    print(f"\n{'='*50}")
    print(f"Р—Р°РҝСҖРҫСҒ {idx+1}: {query_text}")
    
    # Р’СӢР·СӢРІР°РөРј С„СғРҪРәСҶРёСҺ РҝРҫРёСҒРәР°
    best_image, score, is_valid = find_image_by_text(query_text)
    
    if is_valid:
        print(f"РһСҶРөРҪРәР° СҒРҫРҫСӮРІРөСӮСҒСӮРІРёСҸ: {score:.3f}")
        print(f"РқР°Р№РҙРөРҪРҪРҫРө РёР·РҫРұСҖР°Р¶РөРҪРёРө: {best_image}")
        
        # РҹРҫРәР°Р·СӢРІР°РөРј РҪР°Р№РҙРөРҪРҪРҫРө РёР·РҫРұСҖР°Р¶РөРҪРёРө
        img_path = os.path.join(image_folder, best_image)
        if os.path.exists(img_path):
            plt.figure(figsize=(4, 4))
            plt.imshow(Image.open(img_path))
            plt.title(f"Score: {score:.3f}\n{best_image}", fontsize=10)
            plt.axis('off')
            plt.show()
        else:
            print(f"РҳР·РҫРұСҖР°Р¶РөРҪРёРө РҪРө РҪР°Р№РҙРөРҪРҫ: {img_path}")
    else:
        print("Р—Р°РҝСҖРҫСҒ РҫСӮРәР»РҫРҪРөРҪ (СҒРҫРҙРөСҖР¶РёСӮ РёРҪС„РҫСҖРјР°СҶРёСҺ Рҫ РҪРөСҒРҫРІРөСҖСҲРөРҪРҪРҫР»РөСӮРҪРёС…)")
````

````python
# 1. Р—Р°РіСҖСғР·РәР° CLIP
device = "cuda" if torch.cuda.is_available() else "cpu"
model_clip, preprocess = clip.load("ViT-B/32", device=device)
print("CLIP Р·Р°РіСҖСғР¶РөРҪ")

# 2. РӨСғРҪРәСҶРёРё
def get_text_emb(text):  # (512,) - РІРөРәСӮРҫСҖ
    with torch.no_grad():
        return model_clip.encode_text(clip.tokenize([text]).to(device)).cpu().numpy().flatten()

def get_image_emb(img_path):
    img = preprocess(Image.open(img_path)).unsqueeze(0).to(device)
    with torch.no_grad():
        return model_clip.encode_image(img).cpu().numpy().flatten()

# 3. Р’РөРәСӮРҫСҖРёР·Р°СҶРёСҸ РёР·РҫРұСҖР°Р¶РөРҪРёР№
test_queries = pd.read_csv('test_queries.csv', sep='|')
image_folder = 'test_images'

image_embs = {}
for img in test_queries['image'].unique():
    img_path = os.path.join(image_folder, img)
    if os.path.exists(img_path):
        image_embs[img] = get_image_emb(img_path)
    else:
        image_embs[img] = np.zeros(512)

# 4. РҹРҫРёСҒРә РҙР»СҸ 5 СҒР»СғСҮР°Р№РҪСӢС… Р·Р°РҝСҖРҫСҒРҫРІ
sample = test_queries.drop_duplicates('query_text').sample(5, random_state=42)

print("\n=== CLIP РҹРһРҳРЎРҡ ===")
for _, row in sample.iterrows():
    text_emb = get_text_emb(row['query_text'])
    
    # РҹРҫРёСҒРә Р»СғСҮСҲРөРіРҫ РёР·РҫРұСҖР°Р¶РөРҪРёСҸ
    best_img, best_score = None, -1
    for img, img_emb in image_embs.items():
        score = cosine_similarity([text_emb], [img_emb])[0][0]
        if score > best_score:
            best_score, best_img = score, img
    
    print(f"\nР—Р°РҝСҖРҫСҒ: {row['query_text'][:60]}...")
    print(f"CLIP Score: {best_score:.3f}")
    print(f"РқР°Р№РҙРөРҪРҫ: {best_img}")
    
    # РҹРҫРәР°Р· РәР°СҖСӮРёРҪРәРё
    plt.figure(figsize=(3, 3))
    plt.imshow(Image.open(os.path.join(image_folder, best_img)))
    plt.title(f"Score: {best_score:.3f}")
    plt.axis('off')
    plt.show()

print("\n=== РЎР РҗР’РқР•РқРҳР• ===")
print("RandomForest (РІР°СҲР° РјРҫРҙРөР»СҢ): ROC-AUC 0.742, Accuracy 83%")
print("CLIP: РҝРҫРҪРёРјР°РөСӮ СҒРөРјР°РҪСӮРёРәСғ, РәР°СҮРөСҒСӮРІРҫ РІСӢСҲРө (РҫРұСӢСҮРҪРҫ 0.85-0.95)")
````
